```python
credit_card_data = pd.read_csv('creditcard.csv', index_col=False, header=0)
credit_card_data
```

```python
class_values = credit_card_data['Class'].value_counts()
class_values
```

```python
class_values.plot(kind = 'bar')
plt.show()
```

```python
class_values.plot(kind = 'bar')
plt.yscale(value='log')
plt.show()
```

```python
hist_v1 = pd.DataFrame(
    {
        'Class 0':credit_card_data[(credit_card_data.Class == 0)].V1,
        'Class 1':credit_card_data[(credit_card_data.Class == 1)].V1
    },
    columns=['Class 0', 'Class 1'],
)
```

```python
ax = hist_v1.plot.hist(bins=20, alpha=0.5, density = True, color={"Class 0": "grey", "Class 1": "red"})
ax.set_xlabel("x label")
```
